## Central Limit Theorem & Sampling Distribution

The Central Limit Theorem states that sampling distribution of the mean of any independent or random variable will be normal if the sample size is large. So how large is large? According to the rule of thumb, the size should be at least 30. It also requires accuracy. The more closely the sample distribution resemblez a normal distribution the more sample points are required. The closer to the original population resembles the normal population the fewer sample points are required.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Population of size 10,000
population = np.random.binomial(10, 0.4, 10000)
```

```python
sample_means = []
for x in range(1,21):
    sample = np.random.choice(population, size=int(0.30 * len(population)), replace=False) # 30% of population
    sample_means.append(np.mean(sample))
```

```python
# Calculating the mean of sample means and population mean
mean_of_sample_means = np.mean(sample_means)
population_mean = np.mean(population)

print(f'Mean of Sample means: {mean_of_sample_means}')
print(f'Mean of Population: {population_mean}')
```

```python
ax = sns.distplot(sample_means,
                 kde=True,
                 color='orange',
                 hist_kws={'linewidth': 15, 'alpha': 1})
```

```python
# Create the bins and histogram
plt.hist(sample_means, 100, density=True)
```

```python
# Increase sample range from 20 (above) to 100; data is now trending towards a normal distribution.
sample_means = []
for x in range(1,100):
    sample = np.random.choice(population, size=int(0.30 * len(population)), replace=False) # 30% of population
    sample_means.append(np.mean(sample))
    
ax = sns.distplot(sample_means,
                 kde=True,
                 color='orange',
                 hist_kws={'linewidth': 15, 'alpha': 1})
```

```python
# Increase sample range to 10,000...
sample_means = []
for x in range(1,10000):
    sample = np.random.choice(population, size=int(0.30 * len(population)), replace=False) # 30% of population
    sample_means.append(np.mean(sample))
    
ax = sns.distplot(sample_means,
                 kde=True,
                 color='orange',
                 hist_kws={'linewidth': 15, 'alpha': 1})
```
