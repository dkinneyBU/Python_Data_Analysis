I initially created this notebook for the ggplot/ggpy package, but discovered that it is no longer being actively maintained. Having fallen in love with the R ggplot2 syntax, I was motivated to search for an alternative. That led me to **plotnine** which is scary-similar to the R ggplot2 package. 

I converted all the original ggplot/ggpy examples to plotnine in this notebook.

```python
import pandas as pd
from plotnine import *
from plotnine.data import mtcars
```

```python
diamonds = pd.read_csv('../datasets/diamonds.csv',index_col=0)
diamonds.sample(3)
```

```python
(ggplot(diamonds, aes(x='price', color='clarity')) 
 + geom_density() 
 + scale_color_brewer(type='div', palette=7) 
 + facet_wrap('cut'))
```

```python
(ggplot(diamonds, aes(x='carat', y='price', color='clarity')) 
 + geom_point() 
 + scale_color_brewer(type='div'))
```

```python
(ggplot(diamonds, aes(x='carat', y='price', color='clarity')) 
+ geom_point() 
+ xlab("THIS IS AN X LABEL") 
+ ylab("THIS IS A Y LABEL") 
+ ggtitle("THIS IS A TITLE"))
```

```python
ggplot(diamonds, aes(x='carat', y='price')) + geom_point() + facet_wrap(x='clarity', y='cut')
```

```python
ggplot(aes(x='x', y='y', color='z'), data=diamonds.head(1000)) +\
    geom_point() +\
    scale_color_gradient(low='#05D9F6', high='#5011D1')
```
